```python
#Encoding the Categorical data  as we convert into numeric data for a better predictive model
df ['diagnosis'] = df ['diagnosis'].map ({'M':1,'B':0})
df ['diagnosis'].unique()
```

```python
import sklearn
```

```python
#Splitting the data into the Training and Testing set
x = df.drop ('diagnosis',axis=1)
y = df ['diagnosis']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split (x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#Feature Scaling of data
#Using StandardScaler method from Scikitlearn libarary
from sklearn.preprocessing import StandardScaler 
ss = StandardScaler()
x_train = ss.fit_transform (x_train)
x_test = ss.fit_transform (x_test)
```

```python
#Importing Logistic Regression from Scikit learn library
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
#Loading the training data in the model
lr.fit (x_train, y_train)
```

```python
#Predicting output with the test data
y_pred = lr.predict (x_test)
y_pred
```

```python
y_test
```

```python
#accuracy_score() function provided by Scikit-Learn to determine the accuracy rate 
#Accuracy Score of Logistic Regression
from sklearn.metrics import accuracy_score
print("Accuracy Score of Logistic Regression: ",  (accuracy_score (y_test,y_pred)) )
```

```python
#Importing from Decision Tree Classifier from Scikit learn library
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
#Loading the training data in the model
dtc.fit (x_train,y_train)
```

```python
#Predicting output with the test data
y_pred = dtc.predict (x_test)
y_pred
```

```python
#Accuracy Score of Decision Tree Classifier
from sklearn.metrics import accuracy_score
print("Accuracy Score of Decision Tree Classifier : ", (accuracy_score (y_test,y_pred)))
```

```python
#Importing from RandomForest Classifier from Scikit learn Library
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
#Loading the training data in the model
rfc.fit (x_train,y_train)
```

```python
#Predicting output with test data
y_pred = rfc.predict (x_test)
y_pred
```

```python
#Accuracy Score of Random Forest Classifier
from sklearn.metrics import accuracy_score
print("Accuracy Score of Random Forest Classifier : ",(accuracy_score (y_test,y_pred)))
```

```python
#Importing svm from scikit library
#Support Vector Classifier - SVC
from sklearn import svm
svc = svm.SVC ()
#Loading the training data in the model
svc.fit (x_train,y_train)
```

```python
#Predicting output with test data
y_pred = svc.predict (x_test)
y_pred
```

```python
#Accuracy Score of Support vector classifier
from sklearn.metrics import accuracy_score
print(" Accuracy Score of Support vector classifier: ", (accuracy_score (y_test,y_pred )))
```

```python
# From the accuracy and metrics above, the model that performed the best on the test data was the SVC with accuracy
# score of 97%.So let’s choose that model to detect cancer cells in patients.
```
